2. Write a function `arg_first_alphabetically(lst)`, which does the same thing as in exercise 1 but returns the index of the first string alphabetically.

```python
def arg_first_alphabetically(lst):
  unsorted_lst=[]
  for item in lst:
    unsorted_lst.append(item)
  for i in range(len(lst)-1, 0, -1):
    for j in range(i):
      if lst[j] > lst[j+1]:
        lst[j], lst[j+1] = lst[j+1], lst[j]
  for k in range(1, len(unsorted_lst)):
    if unsorted_lst[k]==lst[0]:
      print("Index of the first string alphabetically is:",k)
      break
```

6. Imagine you have a normal distributed random variable `x`. For example `x` can be grades on this exam. Using the normal distribution generator and histogram functions from lecture (provided below) and `cumulative_sum` from previous question to compute what is the value of $x_{90}$ in $\sigma$ such that 90% of the values $x$ are below $x_{90}$. In other words:

$$
\int_{-\infty}^{x_{90}} N(x;\mu=0,\sigma=1) dx = 0.9
$$

```python
import math,random

def arange(x_min,x_max,steps=10):
    step_size=(x_max-x_min)/steps
    x=x_min
    out = list()
    for i in range(steps):
        out.append(x)
        x+=step_size
    return out

def generate_normal(N,m=0,s=1):
    out = list() 
    
    while len(out)<N:
#    for _ in range(int(float(N)/2)+1):
        U1=random.random()
        U2=random.random()
    
        out.append(s*math.sqrt(-2*math.log(U1))*math.cos(2*math.pi*U2)+m)
        out.append(s*math.sqrt(-2*math.log(U1))*math.sin(2*math.pi*U2)+m)

    return out[:N]

def histogram(data, n_bins=1000,x_min=None, x_max=None):
    if x_min==None:
        x_min=min(data)
    if x_max==None:
        x_max=max(data)
        
    bin_edges = arange(x_min,x_max,n_bins)
    bin_edges.append(x_max)

    hist=[0]*n_bins
    
    for d in data:
        for i in range(n_bins):
            if d>=bin_edges[i] and d<bin_edges[i+1]:
                hist[i]+=1
                break
                
    return hist,bin_edges

def cumulative_sum(lst):   #giving out cumulative percentage sum of the normal values
  a=[0]
  p=[]
  for i,j in zip(lst,a):
    a.append(i+j)
  a.pop(0)
  for item in a:
    p.append(item/100000)
  return p
```

```python
normal_lst= generate_normal(100000)
print("Normal Random Variable values:",normal_lst)
h,b=histogram(normal_lst)
c=cumulative_sum(h)
print("Histogram bin edges:",b)
print("frequency within bins:",h)
print("Cumulative Sum List:",c)
#Finding x_90 value
for n,m in zip(b,c):
  if(m>0.9):
    x_90=n
    break
print("The bin value for which probability<0.9 is:",x_90)
```

For Probability=0.9, the z-score is 1.28 approx. This also equals $x_{90}$ when $\mu=0,\sigma=1$. Therefore, $x_{90}$= 1.28$\sigma$ for $\sigma$=1 using Gaussian Distribution.
