    Create a matrix class with the following properties:
        It can be initialized in 2 ways:
            with arguments n and m, the size of the matrix. A newly instanciated matrix will contain all zeros.
            with a list of lists of values. Note that since we are using lists of lists to implement matrices, it is possible that not all rows have the same number of columns. Test explicitly that the matrix is properly specified.
        Matrix instances M can be indexed with M[i][j] and M[i,j].
        Matrix assignment works in 2 ways:
            If M_1 and M_2 are matrix instances M_1=M_2 sets the values of M_1 to those of M_2, if they are the same size. Error otherwise.
            In example above M_2 can be a list of lists of correct size.



```
class matrix:
  #allows for two types of initialization
  def __init__(self, *args):
    #*args is a list of lists of values if the first (and only) argument in
    #the *args tuple is a list --> create matrix of given those values
    if isinstance(args[0],list):
      Pass = True
      #checks if matrix size is valid
      for rows in range(len(args[0])):
        if len(args[0][rows])!=len(args[0][0]):
          Pass = False
          print("Invalid matrix dimensions.")
      if Pass:
        matrix = args[0]
    #*args specifies the size of the matrix if the first argument
    #in the *args tuple is an integer, --> create matrix of zeroes with those dimensions
    elif isinstance(args[0],int): 
      Pass = True 
      matrix = list()
      for rows in range(args[0]):
        templist = [0 for j in range(args[1])]
        matrix.append(templist)
    #only matrix attribute is 'rawmatrix'
    if Pass:
      self.rawmatrix = matrix

  def to_list(self):
    return self.rawmatrix

  #allows for indexing in different ways
  def __getitem__(self, *args):
    #if *args[0] (the only argument) is a tuple, *args[0] is a pair of coordinates --> return specified element
    if isinstance(args[0],tuple):
      return self.to_list()[args[0][0]][args[0][1]]
    #if *args[0] is not a tuple, it's the row coordinate of an element (an integer) --> return that 
    #row so that matrix[i][j] returns an element (or slice)
    else:
      return self.to_list()[args[0]]

  #allows for resetting of the 'rawmatrix' list 
  def __setattr__(self,attribute,value):
    #checks if assignment attempt is to 'rawmatrix', if the value is a list, and if 'rawmatrix' has a list already assigned to it
    #(still allows for 'rawmatrix' initialization by __init__)
    if attribute=="rawmatrix" and (isinstance(value,list)) and (self.__dict__.get(attribute,False)!=False):
      #allows reassignment if 'rawmatrix' and the new list are of the same dimensions
      if (len(self.rawmatrix), len(self.rawmatrix[0]))==(len(value), len(value[0])):
        self.__dict__[attribute]=value
      else:
        print("Incompatible matrix dimensions.")
    else:
      self.__dict__[attribute]=value

  def display(self):
    for i in self.to_list():
      print(i)
    return None
```

Write functions that create special matrices (note these are standalone functions, not member functions of your matrix class):

    constant(n,m,c): returns a n by m matrix filled with floats of value c.
    zeros(n,m) and ones(n,m): return n by m matrices filled with floats of value 0 and 1, respectively.
    eye(n): returns the n by n identity matrix.



```
def constant(m,n,c):
  return [[c for j in range(n)] for i in range(m)]

def zeros(m,n):
  return [[0. for j in range(n)] for i in range(m)]

def ones(m,n):
  return [[1. for j in range(n)] for i in range(m)]

def eye(n): 
  matrix = list()
  for i in range(n):
    temp = list()
    for j in range(n):
      if i==j:
        temp.append(1.)
      else:
        temp.append(0.)
    matrix.append(temp)
  return matrix
```
