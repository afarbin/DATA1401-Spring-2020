````
def win(board,size):
  #player 2 wins rows and columns
  for i in range(size):
    for j in range(size):
      if(board[i][0]==2 and board[i][1]==2 and board[i][2]==2):
        return 2
      elif(board[0][j]==2 and board[1][j]==2 and board[2][j]==2):
        return 2
      elif(board[0][0]==2 and board[1][1]==2 and board[2][2]==2):
        return 2
      elif(board[0][2]==2 and board[1][1]==2 and board[2][1]==2):
        return 2
  #player 1 wins rows and columns
      elif(board[i][0]==1 and board[i][1]==1 and board[i][2]==1):
        return 1
      elif(board[0][j]==1 and board[1][j]==1 and board[2][j]==1):
        return 1
      elif(board[0][0]==1 and board[1][1]==1 and board[2][2]==1):
        return 1
      elif(board[0][2]==1 and board[1][1]==1 and board[2][1]==1):
        return 1

  #if incomplete
  for i in range(0,3):
    if (i==0):
      return -1
  #draw
  return 0

````
