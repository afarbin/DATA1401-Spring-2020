*Exercise 2:* Write a function that takes a `n` by `n` matrix representing a tic-tac-toe game, and returns -1, 0, 1, or 2 indicating the game is incomplete, the game is a draw, player 1 has won, or player 2 has one, respectively. Here are some example inputs you can use to test your code:

````python
#checks if a player has won, lost, or if there is a tie
def status(board, playerx, playero, empty):
  p1comp = list()
  p2comp = list()
  complist = list()
  boardlength = range(len(board))
  #filling comparison lists:
  for i in range(((len(boardlength)-1)/2)):
    p1comp.append(playerx+" ")
    p2comp.append(playero+" ")

  #compares current and win condition board lists:
  def compare(complist, p1comp, p2comp):
    if complist == p1comp:
      print("Player 1 wins!")
      return True
    elif complist == p2comp:
      print("Player 2 wins!")
      return True

  #checks for draw:
  def tie(board, boardlength, empty):
    fullcount = 0
    for rows in boardlength:
      if rows%2!=0:
        for cols in boardlength:
          if board[rows][cols] == playerx + " " or board[rows][cols] == playero + " ":
            fullcount+=1
    if fullcount == (((len(boardlength)-1)/2)**2):
      print("The game is a draw.")
      return True

  #checks for win: 
  #vertical check:
  for cols in range(len(board[1])):
    if cols%2!=0:
      complist = list()
      for rows in boardlength:
        if rows%2!=0:
          if board[rows][cols] == playerx + " " or board[rows][cols] == playero + " ":
            complist.append(board[rows][cols])
      if compare(complist, p1comp, p2comp):
        return True

  #horizontal check:
  for rows in boardlength:
    if rows%2!=0:
      complist = list()
      for cols in boardlength:
        if board[rows][cols] == playerx + " " or board[rows][cols] == playero + " ":
          complist.append(board[rows][cols])
      if compare(complist, p1comp, p2comp):
        return True

  #diagonal check:
  complist=list()
  for n in range(((len(boardlength)-1)/2)):
    if board[2*n+1][2*n+1] == playerx + " " or board[2*n+1][2*n+1] == playero + " ":
      complist.append(board[2*n+1][2*n+1])
  if compare(complist, p1comp, p2comp):
    return True
  #antidiagonal check:
  complist = list()
  for n in range(((len(boardlength)-1)/2)):
    if board[1+2*n][-(2*n+3)] == playerx + " " or board[1+2*n][-(2*n+3)] == playero + " ":
      complist.append(board[1+2*n][-(2*n+3)])
  if compare(complist, p1comp, p2comp):
    return True  
  tie(board, boardlength, empty)
````

*Exercise 8:* Write a function is called with a board and player number, takes input from the player using python's `input`, and modifies the board using your function from exercise 7. Note that you should keep asking for input until you have gotten a valid input that results in a valid move.

````python
def game(board, playernum):
  alphabet = "ABCDEFGHIJKLMNOPQRSTUVWYXZ"
  colnum = (len(board)-1)/2
  end = "Thanks for playing!"
  #creates alphabet grid list to compare against the user's input letter coordinate
  alphlist = []
  for i in range(colnum):
    alphlist.append(alphabet[i])
  #checks for valid player number
  check=0
  while check==0:
    if playernum==1:
      check = 1
    elif playernum==2:
      check = 1
    else:
      playernum = input("Invalid player number. Please re-enter.")
  #accepts and checks letter coordinate:
  check=0
  letcood = input("Please enter your letter coordinate.")
  while check==0:
    for i in range(len(alphlist)):
      if letcood == alphlist[i]:
        letcood = i+1
        check = 1
        break
    if check==0:
      letcood = input("Invalid letter coordinate. Please re-enter.")
  #accepts and checks number coordinate:
  check=0
  numcood = input("Please enter your number coordinate.")
  while check==0:
    if int(numcood)<=colnum and numcood>=1:
      check=1
    else: 
      numcood = input("Invalid number coordinate. Please re-enter.")
  #uses turn to modify the board with the given coordinates
  turn(board, letcood, numcood, playernum, empty)
  if status(board, playerx, playero, empty):
    print(end)
    return None
  #switches players
  if playernum==1:
    playernum=2
  elif playernum==2:
    playernum=1
  #new turn
  return game(board,playernum)
````
