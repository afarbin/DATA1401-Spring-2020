*Exercise 3:* Write a function that takes 2 integers `n` and `m` as input and draws a `n` by `m` game board. For example the following is a 3x3 board:
```
   --- --- --- 
  |   |   |   | 
   --- --- ---  
  |   |   |   | 
   --- --- ---  
  |   |   |   | 
   --- --- --- 
   ```

````python
# Write you solution here
def draw_board(n, m):
  for row in range(n):
    line1=""
    line2=""
    for column in range(m):
      line1+="  ---"
      line2+=" |   "
    line2+="|"
    print(line1)
    print(line2)
    bottom=""
    for column in range(m):
      bottom+="  ---"
    print(bottom)

    
````
